```python
import yfinance as yf

# Загружаем данные о Tesla (тикер TSLA)
tesla = yf.Ticker("TSLA")

# Получаем исторические данные
tesla_data = tesla.history(period="max")

# Сбрасываем индекс
tesla_data.reset_index(inplace=True)

# Отображаем первые пять строк
print(tesla_data.head())
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Получаем HTML со страницы Tesla на macrotrends
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
html_data = requests.get(url).text

# Ищем таблицы
soup = BeautifulSoup(html_data, "html.parser")
tables = soup.find_all("table")

tesla_revenue = None

# Ищем нужную таблицу по тексту "Tesla Quarterly Revenue"
for table in tables:
    if "Tesla Quarterly Revenue" in table.text:
        tesla_revenue = pd.read_html(str(table))[0]
        break

# Проверка, что таблица найдена
if tesla_revenue is None:
    print("❌ Таблица с данными о доходах Tesla не найдена.")
else:
    # Чистим таблицу
    tesla_revenue.dropna(inplace=True)
    tesla_revenue["Tesla Quarterly Revenue"] = tesla_revenue["Tesla Quarterly Revenue"].str.replace("$", "", regex=False)
    tesla_revenue["Tesla Quarterly Revenue"] = tesla_revenue["Tesla Quarterly Revenue"].str.replace(",", "", regex=False)
    tesla_revenue["Tesla Quarterly Revenue"] = pd.to_numeric(tesla_revenue["Tesla Quarterly Revenue"])

    # Показываем последние 5 строк
    print(tesla_revenue.tail())
```

```python
import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(10,6))
    plt.plot(data['Date'], data['Close'], label='Close Price', color='red')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# Предположим, что tesla_data у тебя уже есть и в нем есть колонки 'Date' и 'Close'
# Например, если нет — можно получить данные так:
import yfinance as yf

tesla_data = yf.download('TSLA', start='2020-01-01', end='2025-01-01')
tesla_data.reset_index(inplace=True)

# Строим график
make_graph(tesla_data, "Tesla Stock Price")
```
